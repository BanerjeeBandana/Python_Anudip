```python
#Write a Pandas program to create a Pivot table and find the total sale amount region wise, manager wise, sales man wise.

import numpy as np     #Importing the Numpy library as np
import pandas as pd     #Importing pandas library as pd
df=pd.read_csv("salesdata.csv")     #Reading the csv file
#Creating the Pivot table and finding the total sale amount region wise, manager wise, sales man wise
pd.pivot_table(df,index=['Region','Manager','SalesMan'],values=['Sale_amt'],aggfunc='sum')     
```

```python
#Write a Pandas program to create a Pivot table and find the item wise unit sold.

import numpy as np     #Importing the Numpy library as np
import pandas as pd     #Importing pandas library as pd
df=pd.read_csv("salesdata.csv")     #Reading the csv file
pd.pivot_table(df,index=['Item'],values=['Units'],aggfunc='sum')
```
